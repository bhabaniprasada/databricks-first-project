### **DYNAMIC CAPABILITIES**

```python
dbutils.widgets.text("file_name","")
```

```python
p_file_name = dbutils.widgets.get("file_name")
```

### **DATA READING**

```python
df_orders_autoloader = spark.readStream\
                        .format('cloudFiles')\
                        .option('cloudFiles.format','parquet')\
                        .option('cloudFiles.schemaLocation',f'/Volumes/databricks_ete_project/source/ingestion_metadata/schema/bronze_{p_file_name}')\
                        .load(f'/Volumes/databricks_ete_project/source/raw_data/{p_file_name}/')
                               
                                           
```

### **DATA WRITING**

```python
df_orders_autoloader.writeStream\
    .format('delta')\
    .outputMode('append')\
    .option('checkpointLocation',f'/Volumes/databricks_ete_project/source/ingestion_metadata/checkpoint/bronze_{p_file_name}')\
    .trigger(once=True)\
    .toTable(f'databricks_ete_project.bronze.bronze_{p_file_name}')
```
